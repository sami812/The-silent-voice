```python
import pandas as pd
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
df = pd.read_parquet('train_multilingual.parquet', engine='pyarrow')
print(f"Data Loaded: {len(df)} rows.")
```

```python
df = df[~df['output'].str.contains("<nooutput>", case=False, na=False)]
df = df[df['output'].str.len() < 400]
df = df[df['output'].str.strip() != ""]
print(f"Cleaned Data: {len(df)} rows.")
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1, 3), max_features=50000, min_df=2)
tfidf_matrix = vectorizer.fit_transform(df['instruction'] + " " + df['input'].fillna(''))
```

```python
df.to_parquet('train_multilingual_clean.parquet')
with open('tfidf_vectorizer.pkl', 'wb') as f: pickle.dump(vectorizer, f)
with open('tfidf_matrix.pkl', 'wb') as f: pickle.dump(tfidf_matrix, f)
```
